## timeit

```python
import timeit
```

```python
%timeit np.random.uniform(-90, 90, size=1500)
```

```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < size:
        random_lat = random.randint(-90, 89) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes

%timeit latitudes(1500)
```

#### Using the np.random.uniform(-90.000, 90.000, size=1500) is 100 times faster than using the function,
#### and our code is one line, whereas the function uses eight lines!



# Create Latitude and Longitude Combinations


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a set of random latitude and longitude combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs
```

##  Zip object
- The zip object packs each pair of lats and lngs having the same index in their respective array into a tuple.
- there will be 1,500 tuples of paired latitudes and longitudes
### Unzip
- You can only unzip a zipped tuple once before it is removed from the computer's memory. Make sure you unzip the latitudes and longitudes into the coordinates list before moving on.

```python
# Add the latitudes and longitudes to a list
coordinates = list(lat_lngs)
coordinates[:5]
```

```python
# Use the tuple() function to display the latitude and longitude combinations.
for coordinate in coordinates:
    print(coordinate[0], coordinate[1])
```
